```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset from your folder
data_path = r'C:\Users\Naresh\OneDrive\Desktop\Rooman\ai4i2020_cleaned.csv'
df = pd.read_csv(data_path)

# Set style for seaborn
sns.set(style="whitegrid")

# 1. Plot distribution of 'Target' (machine failure or not)
plt.figure(figsize=(6,4))
sns.countplot(x='Target', data=df, palette='Set2')
plt.title('Distribution of Machine Failure (Target)')
plt.xlabel('Failure (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()

# 2. Correlation heatmap for numerical features
plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Feature Correlation Heatmap')
plt.show()

# 3. Boxplot of sensor readings grouped by failure (Target)
sensor_cols = ['Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']

for col in sensor_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='Target', y=col, data=df, palette='Set3')
    plt.title(f'{col} by Machine Failure (Target)')
    plt.show()

# 4. Pairplot for a few important features with hue by Target
sns.pairplot(df[['Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]', 'Target']], hue='Target', palette='Set1', diag_kind='kde')
plt.suptitle('Pairplot of Sensor Features by Failure', y=1.02)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset from your folder
data_path = r'C:\Users\Naresh\OneDrive\Desktop\Rooman\ai4i2020_cleaned.csv'
df = pd.read_csv(data_path)

# Set style for seaborn
sns.set(style="whitegrid")

# 1. Plot distribution of 'Machine failure'
plt.figure(figsize=(6,4))
sns.countplot(x='Machine failure', data=df, palette='Set2')
plt.title('Distribution of Machine Failure')
plt.xlabel('Failure (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()

# 2. Correlation heatmap for numerical features
plt.figure(figsize=(12,10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Feature Correlation Heatmap')
plt.show()

# 3. Boxplots of sensor data grouped by 'Machine failure'
sensor_cols = ['Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']

for col in sensor_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='Machine failure', y=col, data=df, palette='Set3')
    plt.title(f'{col} by Machine Failure')
    plt.show()

# 4. Pairplot of selected features with hue='Machine failure'
sns.pairplot(df[['Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]', 'Machine failure']],
             hue='Machine failure', palette='Set1', diag_kind='kde')
plt.suptitle('Pairplot of Sensor Features by Failure', y=1.02)
plt.show()
```
